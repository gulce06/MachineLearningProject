```python
#filter 
import pandas as pd 
df= pd.DataFrame({"gruplar": ["A","B","C","A","B", "C"],
                 "degisken1": [10,20,30,10,20,50],
                 "degisken2": [50,30,40,50,60,90]},
                columns= ["gruplar","degisken1","degisken2"])
df
```

```python
def filter_func ( x ):
    return x["degisken1"].std()>9
```

```python
df.groupby("gruplar").std()
```

```python
df.groupby("gruplar").filter(filter_func)
```

```python
import pandas as pd 
df= pd.DataFrame({"gruplar": ["A","B","C","A","B", "C"],
                 "degisken1": [10,20,30,10,20,50],
                 "degisken2": [50,30,40,50,60,90]},
                columns= ["gruplar","degisken1","degisken2"])
df
```

```python
df_a = df.iloc[:,1:3]
df_a
```

```python
df_a.transform(lambda x:(x-x.mean())/ x.std())
```

```python
df_a.transform(lambda y: y*y.max())
```

```python
df_a.transform(lambda z: z+z.std()-10)
```
