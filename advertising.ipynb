# Stats Model ile Modelleme

```python
import statsmodels.api as sm
```

```python
X= df[["TV"]]
X[0:5]
```

```python
X=sm.add_constant(X)
X[0:5]
```

```python
y= df["sales"]
y[0:5]
```

```python
lm=sm.OLS(y,X) 
```

```python
model=lm.fit()
```

```python
model.summary()
```

```python
import statsmodels.formula.api as smf
lm = smf.ols("sales ~ TV", df)
model = lm.fit()
model.summary()
```

```python
model.params
# Modelin parametreleri. Beta0 ve Beta1 değerleri.
```

```python
model.summary().tables[1]
```

```python
model.conf_int()
```

```python
model.f_pvalue
```

```python
print ( "Pvalue: ","%.4f" % model.f_pvalue)
```

```python
print ( "F value: ","%.4f" % model.fvalue)
```

```python
print ( "T value: ","%.4f" % model.tvalues[0:1])
```

```python
model.mse_model 
```

```python
model.rsquared
```

```python
model.rsquared_adj
```
